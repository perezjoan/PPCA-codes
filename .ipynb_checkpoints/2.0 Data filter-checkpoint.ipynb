```python
'''
PPCA 2.0 GHS AND OSM DATA CLEANING / FILTER

Author : Perez, Joan

The script processes and filter spatial data from OSM and GHS sources for further analysis. It performs three main tasks: (1) filters out buildings with a 
footprint area less than 15 mÂ², underground buildings, and optionally filters out buildings that have no walls, if the 'wall' column exists. (2) It reads 
and filters Global Human Settlement (GHS) population data by rounding values and removing meshes with zero population (3) It filters OpenStreetMap (OSM) 
streets data to separate pedestrian and non-pedestrian streets based on the following attributes: "motorway|motorway_link|trunk|trunk_link|cycleway"
(4) It filters OSM land use data to identify non-populated areas based on the following attributes:"construction|cemetery|education|healthcare|industrial|
military|railway|religious|port|winter_sports". Filtered and filtered out features are saved in two separate geopackage files.

Requirements:
- A specific working environment (see README on the github page of the project https://github.com/perezjoan/PPCA-codes?tab=readme-ov-file)
- Output file from PPCA 1.0 ('ghs_{date}_vector'(Polygon),  GHS population data at a given date ; 'osm_all_area_categories ' (Polygon), OSM land use
data with non-populated areas ; 'osm_all_streets' (LineString), OSM all streets)

Guide to run the script:
- Fill 0.2 box

Outputs :
- A first geopackage file with 4 layers :
    * 'osm_building_filtered' (Polygon), OSM buildings with light structures filtered out
    * 'ghs_populated_2020_vector'(Polygon),  GHS population data with non null values
    * 'osm_non_populated_areas' (Polygon), OSM land use data with non-populated areas
    * 'pedestrian_streets' (LineString), OSM pedestrian streets
- A second geopackage file with 3 layers (filtered out features) :
    * 'non_pedestrian_streets' (LineString), OSM non-pedestrian streets
    * 'osm_building_filtered_out' (Polygon), filtered out OSM buildings
    * 'osm_populated_areas' (Polygon), OSM land use data with populated areas

Acknowledgement: This resource was produced within the emc2 project, which is funded by ANR (France), FFG (Austria), MUR (Italy) and 
Vinnova (Sweden) under the Driving Urban Transition Partnership, which has been co-funded by the European Commission.

License: Attribution-ShareAlike 4.0 International - CC-BY-SA-4.0 license
'''
```

```python
# 0.1 : Libraries
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 0.2 : Box to fil with information

# Name of the case study
Name = 'Goteborg'

# Define projected CRS
projected_crs = 'EPSG:3152'

# Recall the ghs date (extracted in previous step)
ghs_date = 2020
```

```python
# 0.3 Data preparation

gpkg = f'PPCA1_{Name}.gpkg'

# GHS
ghs_vector = gpd.read_file(gpkg, layer = f'ghs_{ghs_date}_vector')

#OSM streets
linestrings_gdf = gpd.read_file(gpkg, layer = 'osm_all_streets')

# OSM land use
all_areas = gpd.read_file(gpkg, layer = 'osm_all_area_categories')

# OSM buildings
all_building = gpd.read_file(gpkg, layer = 'osm_all_buildings')
```

```python
# 1. FILTER SMALL BUILDINGS, BUILDINGS WITH NO WALLS & UNDERGROUND BUILDINGS

# Projected CRS to calculate surface area
all_building = all_building.to_crs(projected_crs)
all_building['A'] = all_building.geometry.area

# Create a copy of the original data for the removed buildings
removed_buildings = all_building.copy()

# Filter out buildings with a surface footprint less than 15 mÂ²
building_filtered = all_building[all_building['A'] >= 15]

# Filter out buildings with no walls (if the wall column exists)
if 'wall' in building_filtered.columns:
    building_filtered = building_filtered[building_filtered['wall'] != 'no']

# Remove buildings that have a 0 value in the buildings:levels column (0 = underground buildings)
if 'buildings:levels' in building_filtered.columns:
    building_filtered = building_filtered[building_filtered['buildings:levels'] != 0]

# Determine the removed buildings by comparing the original and filtered GeoDataFrames
removed_buildings = removed_buildings[~removed_buildings.index.isin(building_filtered.index)]

# Plot the removed buildings
fig, ax = plt.subplots(figsize=(10, 10))
removed_buildings.plot(ax=ax, color='red', edgecolor='black', alpha=0.6)

# Adding titles and labels
plt.title('Removed Buildings')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Show the plot
plt.show()
```

```python
# 2. FILTER GHS DATA

# Round the population values in the "values" column
ghs_vector['VALUE'] = ghs_vector['VALUE'].round()

# Filter out rows with a value of 0
ghs_vector = ghs_vector[ghs_vector['VALUE'] != 0]

# Plot
num_observations = len(ghs_vector)
fig, ax = plt.subplots(figsize=(10, 10))
ghs_vector.plot(ax=ax)
ax.set_title(f'Filtered GHS (no null) - number of observations: {num_observations}')
plt.show()
```

```python
# 3. FILTER OSM STREETS DATA
# Attribute values: https://wiki.openstreetmap.org/wiki/Key:highway

# Define maxspeed values to remove
maxspeed_values_to_remove = ['60', '70', '80', '90', '100']

# Remove non-pedestrian streets and streets with specific maxspeed values
removed_streets = linestrings_gdf[
    linestrings_gdf['highway'].str.contains('motorway|motorway_link|trunk|trunk_link|cycleway|primary_link|busway', na=False) |
    (linestrings_gdf['tunnel'] == 'yes') |
    linestrings_gdf['maxspeed'].str.contains('|'.join(maxspeed_values_to_remove), na=False)
]

# Keep pedestrian streets
pedestrian_streets = linestrings_gdf[
    ~linestrings_gdf['highway'].str.contains('motorway|motorway_link|trunk|trunk_link|cycleway|primary_link|busway', na=False) &
    (linestrings_gdf['tunnel'] != 'yes') &
    ~linestrings_gdf['maxspeed'].str.contains('|'.join(maxspeed_values_to_remove), na=False)
]

# Plot pedestrian_streets & non-pedestrian streets
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
num_observations = len(pedestrian_streets)
pedestrian_streets.plot(ax=axes[0], color='blue')
axes[0].set_title(f'Pedestrian streets - number of observations: {num_observations}')
num_observations = len(removed_streets)
removed_streets.plot(ax=axes[1], color='green')
axes[1].set_title(f'Filtered streets - number of observations: {num_observations}')
plt.show()
```

```python
# 4. FILTER OSM LANDUSE DATA

# Remove non-populated landuse areas
# Attribute values: https://wiki.openstreetmap.org/wiki/Key:landuse
non_populated_areas = all_areas[
    all_areas['landuse'].str.contains(
        'construction|cemetery|education|healthcare|industrial|military|railway|religious|port|winter_sports', 
        na=False
    )
]

# Extract populated landuse areas by excluding the non-populated categories
populated_areas = all_areas[~all_areas.index.isin(non_populated_areas.index)]

# Plot
num_observations = len(non_populated_areas)
fig, ax = plt.subplots(figsize=(10, 10))
non_populated_areas.plot(ax=ax)
ax.set_title(f'Filtered GHS (no null) - number of observations: {num_observations}')
plt.show()
```

```python
# A1. Save Outputs
gpkg = f'PPCA2_{Name}.gpkg'
building_filtered.to_file(gpkg, layer='osm_building_filtered', driver="GPKG")
ghs_vector.to_file(gpkg, layer = f'ghs_populated_{ghs_date}_vector', driver="GPKG")
pedestrian_streets.to_file(gpkg, layer='pedestrian_streets', driver="GPKG")
non_populated_areas.to_file(gpkg, layer='osm_non_populated_areas', driver="GPKG")
gpkg = f'PPCA2_{Name}_filtered_out.gpkg'
removed_streets.to_file(gpkg, layer='non_pedestrian_streets', driver="GPKG")
removed_buildings.to_file(gpkg, layer='osm_building_filtered_out', driver="GPKG")
populated_areas.to_file(gpkg, layer='osm_populated_areas', driver="GPKG")
```
